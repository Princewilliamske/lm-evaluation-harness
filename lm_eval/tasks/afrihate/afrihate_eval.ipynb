```
from datasets import get_dataset_config_names, load_dataset, DatasetDict, concatenate_datasets

# Step 1: List all available language configs
configs = get_dataset_config_names("afrihate/afrihate")
print("Available configs on HuggingFace:", configs)
```

```
# Step 2: Load all datasets
datasets_dict = {}
for lang in configs:
    print(f"Loading dataset for language: {lang}")
    try:
        ds = load_dataset("afrihate/afrihate", name=lang)
        datasets_dict[lang] = ds
    except FileNotFoundError:
        print(f"Skipping language '{lang}' due to missing data files.")
    except Exception as e:
        print(f"An unexpected error occurred while loading language '{lang}': {e}")
```

```


# Step 3: Merge datasets across all languages
# We will merge splits individually (train, validation, test)
merged_splits = {}
for split in ['train', 'validation', 'test']:
    split_datasets = []
    for lang, ds in datasets_dict.items():
        if split in ds:
            # Add a language column to track the source language
            ds_split = ds[split].add_column("language", [lang] * len(ds[split]))
            split_datasets.append(ds_split)
    if split_datasets:
        merged_splits[split] = concatenate_datasets(split_datasets)

# Step 4: Create a DatasetDict for the multilingual dataset
all_dataset = DatasetDict(merged_splits)

```

```
from pathlib import Path

# Step 5: Quick check
print("\nAll dataset splits and sizes:")
for split, ds in all_dataset.items():
    print(f"{split}: {len(ds)} examples, labels: {set(ds['label'])}")

# Step 6: Save for lm-eval usage in lm_eval/tasks/afrihate/
save_path = Path("lm_eval/tasks/afrihate/afrihate_all_hate")
save_path.parent.mkdir(parents=True, exist_ok=True)  # ensure folder exists

all_dataset.save_to_disk(save_path)
print(f"\nAll dataset saved to '{save_path}/' folder for lm-eval usage")
```
